```python
import qiskit
import qiskit_aer
```

```python
qreg = qiskit.QuantumRegister(4, 'bob')
qreg2 = qiskit.QuantumRegister(4, 'alice')
creg = qiskit.ClassicalRegister(4, 'eve')
creg2 = qiskit.ClassicalRegister(4, 'eve2')

quantum_circuit = qiskit.QuantumCircuit(qreg, qreg2,creg,creg2)
quantum_circuit.h(0)
quantum_circuit.cx(0, 1)
quantum_circuit.measure(qreg, creg)
quantum_circuit.measure(qreg2, creg2)

quantum_circuit.draw('mpl')
```

```python
backend = qiskit_aer.Aer.get_backend('qasm_simulator')
job = backend.run(quantum_circuit)

job.result().get_counts()
```
